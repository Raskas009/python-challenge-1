```python
# Full menu dictionary with nested categories and items
menu = {
    "Snacks": {
        "Cookie": 0.99,
        "Banana": 0.69,
        "Apple": 0.49,
        "Granola bar": 1.99
    },
    "Meals": {
        "Burrito": 4.49,
        "Teriyaki Chicken": 9.99,
        "Sushi": 7.49,
        "Pad Thai": 6.99,
        "Pizza": {
            "Cheese": 8.99,
            "Pepperoni": 10.99,
            "Vegetarian": 9.99
        },
        "Burger": {
            "Chicken": 7.49,
            "Beef": 8.49
        }
    },
    "Drinks": {
        "Soda": {
            "Small": 1.99,
            "Medium": 2.49,
            "Large": 2.99
        },
        "Tea": {
            "Green": 2.49,
            "Thai iced": 3.99,
            "Irish breakfast": 2.49
        },
        "Coffee": {
            "Espresso": 2.99,
            "Flat white": 2.99,
            "Iced": 3.49
        }
    },
    "Dessert": {
        "Chocolate lava cake": 10.99,
        "Cheesecake": {
            "New York": 4.99,
            "Strawberry": 6.49
        },
        "Australian Pavlova": 9.99,
        "Rice pudding": 4.99,
        "Fried banana": 4.49
    }
}

# Step 1: Initialize the order list to store customer selections
order_list = []

# Welcome the customer
print("Welcome to Rajs Bomb food truck!")

# Main ordering loop for menu
place_order = True
while place_order:
    # Display the main menu categories
    print("\nMenu Categories:")
    i = 1
    menu_items = {}
    for category in menu.keys():
        print(f"{i}: {category}")
        menu_items[i] = category
        i += 1

    # Get the customer's category selection
    menu_category = input("Please enter the menu category number: ")

    if menu_category.isdigit():
        menu_category = int(menu_category)
        if menu_category in menu_items:
            selected_category = menu_items[menu_category]
            print(f"\nYou selected: {selected_category}")

            # Display the items within the selected category
            i = 1
            item_options = {}
            print("Item # | Item Name                     | Price")
            print("-------|-------------------------------|-------")
            for key, value in menu[selected_category].items():
                if isinstance(value, dict):  # Handle nested sub-items (like different types of pizza)
                    for sub_key, sub_value in value.items():
                        item_name = f"{key} - {sub_key}"
                        item_options[i] = {"Item name": item_name, "Price": sub_value}
                        print(f"{i:<7}| {item_name:<30} | ${sub_value:.2f}")
                        i += 1
                else:
                    item_options[i] = {"Item name": key, "Price": value}
                    print(f"{i:<7}| {key:<30} | ${value:.2f}")
                    i += 1

            # Get the customer's item selection
            item_selection = input("\nPlease enter the item number: ")

            if item_selection.isdigit():
                item_selection = int(item_selection)
                if item_selection in item_options:
                    selected_item = item_options[item_selection]["Item name"]
                    selected_price = item_options[item_selection]["Price"]

                    # Get the quantity
                    quantity = input(f"How many {selected_item}s would you like to order? (Default: 1): ")
                    if not quantity.isdigit():
                        print("Invalid quantity! Defaulting to 1.")
                        quantity = 1
                    else:
                        quantity = int(quantity)

                    # Add the selected item to the order list
                    order_list.append({
                        "Item name": selected_item,
                        "Price": selected_price,
                        "Quantity": quantity
                    })
                    print(f"Added {quantity} {selected_item}(s) to your order.")

                else:
                    print("Invalid item selection.")
            else:
                print("Please enter a valid item number.")
        else:
            print("Please select a valid menu category number.")
    else:
        print("You didn't select a valid number.")

    # Ask the customer if they want to continue ordering
    while True:
        keep_ordering = input("\nWould you like to keep ordering? (Y/N): ").lower()
        match keep_ordering:
            case "y":
                break
            case "n":
                place_order = False
                print("\nThank you for your order!")
                break
            case _:
                print("Invalid input! Please enter 'Y' or 'N'.")

# Display the order receipt
print("\nYour Order Receipt:")
print("Item Name                     | Price  | Quantity")
print("------------------------------|--------|----------")
for item in order_list:
    item_name = item["Item name"]
    price = f"${item['Price']:.2f}"
    quantity = item["Quantity"]

    # Calculate spacing for the receipt formatting
    item_spacing = " " * (30 - len(item_name))
    price_spacing = " " * (8 - len(price))

    # Print the formatted receipt line
    print(f"{item_name}{item_spacing}|{price}{price_spacing}| {quantity}")

# Calculate and display the total price
total_price = sum(item["Price"] * item["Quantity"] for item in order_list)
print(f"\nTotal Price: ${total_price:.2f}")
```
